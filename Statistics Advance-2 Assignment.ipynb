Question 4: Under what conditions is the binomial distribution used, and how does it relate to the Bernoulli
distribution?

Conditions for Using the Binomial Distribution


1. Fixed Number of Trials (n): The experiment is repeated a set number of times.

Example: Rolling a die 10 times.
2. Binary Outcomes: Each trial results in one of two outcomes: success or failure.

Example: Getting heads or tails in a coin toss.

3. Constant Probability of Success (p): The probability of success remains the same for each trial.

Example: The probability of flipping heads with a fair coin is always 0.5.
4. Independence: Each trial is independent of the others. The outcome of one trial does not affect the outcome of another.

Example: The result of one die roll doesn‚Äôt influence subsequent rolls.

p.

Question6: Define the terms "probability distribution" and "probability density function" (PDF). How does a
PDF differ from a probability mass function (PMF)?




Probability Distribution

A probability distribution describes how the probabilities of different outcomes are distributed for a random variable. It shows the likelihood of each possible value or range of values that a random variable can take.

For Discrete Variables: Describes the probabilities of individual values (e.g., outcomes of rolling a die).

For Continuous Variables: Describes the probability of the random variable falling within a certain range (e.g., a person‚Äôs height).

Probability Density Function (PDF)
A probability density function (PDF) is used to describe the probability distribution of a continuous random variable. It gives the relative likelihood that the variable will take a particular value within a range.

The PDF itself does not give the probability for specific values but instead the probability of the variable falling within a range.
The area under the PDF curve over an interval represents the probability that the variable falls within that interval.


Difference Between PDF and PMF

Probability Mass Function (PMF):Used for discrete random variables.

Gives the probability of the random variable taking a specific value.
Example: For a die roll,
P(X=2)=1/6.


Probability Density Function (PDF):
Used for continuous random variables.

Gives the relative likelihood of the variable being near a particular value. The actual probability of taking any specific value is zero because the variable can take infinitely many values within a range.
Example: The probability of a person‚Äôs height being exactly 170 cm is 0, but the PDF helps calculate the probability of the height falling within a range (e.g., between 170 cm and 175 cm).


Question7: Explain the Central Limit Theorem (CLT) with example.




Central limit theorem

When a large enough sample size is drawn from any population with a finite mean and variance, the sampling distribution of the sample mean will approach a normal distribution, regardless of the shape of the original population distribution.

Key Points of the CLT:

1. Sample Size: The larger the sample size, the more the sampling distribution of the sample mean will resemble a normal distribution. A common rule of thumb is a sample size of 30 or more is sufficient, although this can vary depending on the original population distribution.

2. Population Distribution: The original population can be of any distribution (e.g., skewed, uniform, or even binomial). As long as you take sufficiently large random samples, the distribution of the sample means will approximate a normal distribution.

3. Mean and Variance:

The mean of the sampling distribution will be approximately equal to the population mean (Œº).

The standard deviation of the sampling distribution (known as the standard error) will be the population standard deviation (œÉ) divided by the square root of the sample size (n)




Example of Central Limit Theorem


Population: Suppose we have a population of exam scores that follow a right-skewed distribution, with most students scoring low and a few scoring very high.

Sample: Now, we randomly draw samples of size
n=50 from this population and calculate the mean score for each sample.

Result:

Even though the population distribution of exam scores is skewed, according to the CLT, the sampling distribution of the sample means (i.e., the distribution of all possible sample means from the population) will tend to follow a normal distribution as the sample size increases.
If we keep taking different random samples of size 50, plot their means, and repeat this many times, we‚Äôll observe that the distribution of those means forms a bell-shaped curve, i.e., a normal distribution.

Question9: Given a sample mean of 105, a population mean of 100, a standard deviation of 15, and a sample
size of 25, calculate the z-score and p-value. Based on a significance level of 0.05, do you reject or fail to
reject the null hypothesis?

 Task: Write Python code to calculate the z-score and p-value for the given data.

Objective: Apply the formula for the z-score and interpret the p-value for hypothesis testing.

```python
from scipy import stats

#Given data
sample_mean = 105
population_mean = 100
std_dev = 15
sample_size = 25

#z score calculation
z_score = (sample_mean - population_mean) / (std_dev / (sample_size ** 0.5))

#p value calculation
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

#Significance level
alpha = 0.05

# Determine whether to reject the null hypothesis
reject_null = p_value < alpha

z_score, p_value, reject_null
```

Conclusion:

Since the p-value is greater than the significance level of 0.05, we fail to reject the null hypothesis.

This indicates that there is not enough evidence to suggest that the sample mean of 105 is significantly different from the population mean of 100.

Question10: Simulate a binomial distribution with 10 trials and a probability of success of 0.6 using Python.
Generate 1,000 samples and plot the distribution. What is the expected mean and variance?

Task: Use Python to generate the data, plot the distribution, and calculate the mean and variance.

Objective: Understand the properties of a binomial distribution and verify them through simulation.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

n_trials = 10
p_success = 0.6
n_samples = 1000

# Simulate binomial distribution
samples = np.random.binomial(n_trials, n_success, n_samples)

#Calculate mean and variance
expected_mean = n_trials * p_success
expected_variance = n_trials * p_success * (1 - p_success)

# Plot the distribution
sns.histplot(samples,bins=np.arange(-0.5, n_trials + 1.5, 1), kde=False, stat = 'density')
plt.title('Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.xticks(range(n_trials + 1))
plt.grid()
plt.show()
```

```python
expected_mean, expected_variance
```

Interpretation

The simulated binomial distribution reflects the theoretical properties:

The mean of the distribution is 6.0, indicating that, on average, we expect 6 successes in 10 trials.

The variance of the distribution is 2.4, which indicates the degree of variability around the mean.
